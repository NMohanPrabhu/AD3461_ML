<b>Algorithm: </b>
<ol>
<li>
Import necessary Library files </li>
<li>Load the iris data set </li>
<li>The data is split into features (X) and target (y).X contains feature information of iris data 
set (sepal length, sepal width, petal length, petal width) and y contains iris class (Iris 
setosa, Iris virginica and Iris versicolor) labels </li>
<li>The data is split into training and test sets using the train_test_split function from the 
sklearn.model_selection module. </li>
<li>Create an instance of Decision Tree classfier </li>
<li>Fit the classifier to the training data X_train and y_train </li>
<li>Perform the prediction </li>
<li>Estimate the performance metrics such as accuracy, precision and recall for the classifier. </li>
<li>Print the Decision tree and predict classification label for a given instance</li> </ol>

program:


```python
import pandas as pd 
from sklearn import metrics 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import precision_score 
from sklearn.metrics import recall_score 
from sklearn.metrics import confusion_matrix 
from sklearn.datasets import load_iris 
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
```

```python
iris = load_iris() 
X = iris.data 
y = iris.target 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=4) 
# Create Decision Tree classifer object 
clf = DecisionTreeClassifier() 
clf.fit(X_train, y_train) 
y_pred = clf.predict(X_test) 
```

```python
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred)) 
accuracy = metrics.accuracy_score(y_test, y_pred) 
precision = metrics.precision_score(y_test, y_pred,average='micro') 
recall = metrics.recall_score(y_test, y_pred,average='micro')
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 
 
class_report = classification_report(y_test, y_pred) 
print("Classification Report:\n", class_report) 
```

```python
from sklearn import tree 
from matplotlib import pyplot as plt 
fig = plt.figure(figsize=(25,20)) 
_ = tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names,         
filled=True) 
print(clf.predict([[5.1, 3.4, 1.3, 0.1]])) 
```
