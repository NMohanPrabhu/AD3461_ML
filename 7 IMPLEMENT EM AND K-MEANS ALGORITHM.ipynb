<b>Algorithm:</b> 
1. Import necessary Library files 
2. Load the iris data set 
3. The data is split into features (X) and target (y).X contains feature information of iris data set 
(sepal length, sepal width, petal length, petal width) and y contains iris class (Iris setosa, Iris 
virginica and Iris versicolor) labels 
4. The data is split into training and test sets using the train_test_split function from the 
sklearn.model_selection module. 
5. Create an instance of Kmeans classfier 
6. Fit the classifier to the training data X_train and y_train 
7. Predict the accuracy of the classifier 
8. Using the same data set create an instance of Gaussian mixture  
9. Fit the classifier to the training data X_train and y_train 
10. Predict the accuracy of the classifier 
11. Plot the cluster identified by Kmeans and Gaussian mixture model.

Program: 

```python
from sklearn.mixture import GaussianMixture 
import sklearn.metrics as metrics 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn import datasets 
import pandas as pd 
import numpy as np 
from sklearn.metrics import accuracy_score 
%matplotlib inline 
```

```python
iris = datasets.load_iris() 
X = pd.DataFrame(iris.data) 
X.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'] 
y = pd.DataFrame(iris.target) 
y.columns = ['Targets'] 
model=KMeans(n_clusters=3, random_state=0).fit(X) 
print('The accuracy score of K-Mean: ',metrics.accuracy_score(y, model.labels_)) 
#The accuracy score of K-Mean:  0.24
```

```python
#Another Way
kmeans = KMeans(n_clusters=3,random_state=0) 
clusters = kmeans.fit_predict(X) 
acc = accuracy_score(iris.target, clusters) 
acc 
```

```python
gmm=GaussianMixture(n_components=3, random_state=0).fit(X) 
y_cluster_gmm=gmm.predict(X) 
print('The accuracy score of EM: ',metrics.accuracy_score(y, y_cluster_gmm)) 
```
